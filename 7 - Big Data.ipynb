### **Big Data Nedir?**

Sozluk anlami olarak _Buyuk Veri, geleneksel veri isleme yontemleriyle ele alinamayacak kadar buyuk ver karmasik olan veri kumelerini_ analiz etme, sistematik olarak bilgi cikarma veya bu kumelerle farkli yollarla ilgilenen bir alandir. <br>
Kisaca, cesitli kaynaklardan elde edilen devasa buyuklukteki verilerin gelismis yazilimlar aracigiyla anlamdirilmasi ve kullanilmasidir.

### **Big Data Ozellikleri**



Herhangi bir mecra uzerinden toplanman verilerin Buyuk Veri olarak siniflandirilabilmesi icin cesitli ozelliklere sahip olmasi gerekmektedir. Ingilizcede 5V olarak tanimlanan ve daha bircok tanimlamasi bulunan bu ozellikleri 5 madde altinda inceleyecek olursak;
* **Velocity - Hiz**
    * Veri yuksek hizlarda ve surekli uretiliyor olmali.
* **Volume - Hacim**
    * Veri akisi surekli ve buyuk hacimlerde olmali.
* **Value - Deger**
    * Toplanan verilerden anlamli bilgi edinilebilmesi.
* **Variety - Cesitlilik**
    * Farkli turden veriler icermeli.
* **Veracity - Dogruluk**
    * Guvenilir ve dogru bilgiler icermeli.
    
<br>

_Yaygin olan yanilgilardan bir tanesi de bir veri kumesine buyuk veri denilebilmesi icin bu ozelliklerin hepsini barindirmasi gerektigidir. Eger bir veri kumesi iliskisel veri tabanlarinda veya daha genl tabirle geleneksel yontemlerle islenemiyorsa o veri kumesine buyuk veri denir._

### **Apache Hadoop**



Apache Hadoop, büyük miktarda veri ve hesaplama içeren sorunları çözmek için birçok bilgisayardan oluşan bir ağın kullanılmasını kolaylaştıran bir açık kaynaklı yazılım araçları koleksiyonudur.
* Buyuk veri teknolojilerinin temelini olusturur.
* Geleneksel ontemler ile etkin olarak islenmesi mumkun olmayan verilerin islenebilmesine olanak saglamaktadir.
* Bir bilgisayar kumesinin belirli bir isi yapmak icin tek bir bilgisayar gibi birlikte hareket etmesini saglamaktadir.
<br>

#### **Hadoop Bilesenleri**
* **Hadoop Common:**
    * Tum Hadoop bilesenlerini destekleyen ortak gereksinimleri barindirmaktadir.
* **Hadoop Distributed File System (HDFS)**
    * 
    * NTFS, FAT32,ext4 gibi bir dosya sistemidir.
    * Bu dosya sistemlerinden farki buyuk boyutlardaki verileri dagitik bir sekilde depolamayi ve kontrol etmeyi saglamasidir.
* **Hadoop YARN**
    * Kaynak yonetimi ve is planlamasi icin kullanilmaktadir.
    * Birden fazla makine uzerindeki kullanimda etkin rol oynamaktadir.
* **Hadoop MapReduce**
    * Ayni agdaki dagitik bilgisayar kumeleri uzerinde buyuk veri islemlerinin gerceklestirilebilmesi icin gelistirilmis bir programlama modelidir.

<br>

#### **Hadoop Kume Yapisi**

* Master Node olarak nitelendirilmis ana makine, Slave Nodes olarak belirtilen alt makinelere is dagitimi ve gorevlendirme gibi bircok gorevi yerine getirmektedir. Boylece dagitik bilgisayar kumeleri duzgun bir sekilde calisabilmektedir.

<br>

#### **MapReduce**
Bilgisayar toplulugunun belirli bir amac icin birlikte hareket edebilmesini saglayan programlama modeli/yazilimdir.
<br>


* Verilen metin icerisindeki elemanlari sayma islemini MapReduce ile orneklendirmesi yukarida verilmistir.

#### **Hadoop Eksileri**
* Dis tabanli calisan bir modeldir.
* Her MapReduce gorevinde diskten okuma ve diske yazma islemi yapilir.
* Veri Madenciligi veya Makine Ogrenmesi gibi iteratif islemler zaman alir ve kaynaklari mesgul eder.
* 
* Yukarida goruldugu gibi her iterasyonda diske yazip diskten okuma islemlerinden dolayi makine ogrenmesi islemlerinde bir problem yaratmistir.
* Bu probleme cozum olarak, her seferinde diske gidip gelmek yerine gecici hafiza olarak in memory RAM kullanan ve Apache Hadoop uzerine kurulmus ve iteratif islemlerde bizlere daha iyi imkanlar sunan **Apache Spark** ortaya cikmistir.

### **Apache Spark**

* Spark Core, Hadoop'un bagli oldugu MapReduce'a gore potansiyel olarak daha hghizli ve esnek bir _alternatif_ olarak tasarlanmistir. 
* Gunumuzde daha fazla tercih edilen Spark'in MapReduce'dan 100 kat daha hizli calistigi iddia edilmektedir.
* Spark ve MapReduce arasindaki en temel fark, Spark'in verileri sonraki adimlar icin islemesi ve gecici bellekte tutmasidir. Bu sebeple surekli diskte yazma/okuma maliyetinden kurtulmus ve daha yuksek hizlarda islem yapilabilmesine olanak saglanmistir.
* Java, Scala, Python ve R ile uygulama gelistirilebilmektedir.
* Genellestiricidir: Spark SQL, Spark MLlib, Spark Streaming, GraphX ayni uygulamada kullanilabilir.
#### **Spark Calisma Prensibi**
* Apache Spark, HDFS, NoSQL veritabanlari ve Apache Hive gibi cesitli veri havuzlarindaki verileri isler. 
* Spark, buyuk veri analizi uygulamalarinin performansini arttirmak icin bellekici islemeyi desteklemektedir. Ancak, veri kumeleri mevcut sistem bellegine sigamayacak kadar buyuk oldugunda ise disk tabanli islemeyi de gerceklestirebilmektedir.
* Spark, hiyerarsik bir mimariye sahiptir. Spark surucusu, calisan dugumleri yoneten ve kume ytoneticisini kontrol eden ana dugumdur.
* Spark Core motoru, temel veri turu olarak esnek dagitilmis veri kumelerini (RDD) kullanmaktadir.
* RDD, hesaplama karmasikligini kullanicilardan gizleyecek sekilde tasarlanmis bir yapiya sahiptir.
* Spark Core, verileri toplar ve daha sonra hesaplanabilecegi farkli bir veri doisuna tasir, Sonrasinda ise analitik model uzerinden calistirilabilecegi bir sunucu kumesine boler. Kullanicinin, belirli dosyalarin nereye gonderilecegini, dosyalari depolamak veya almak icin hangi hesaplama kaynaklarinin kullanilacagini tanimlamasi _gerekli degildir._
#### **Resilient Distributed Datasets (RDD)**
* RDD, kumelerdeki birden cok dugum arasinda dagitilabilen ve paralel olarak uzerinde calisilabilen, hataya dayanikli oge koleksiyonlaridir.
* RDD, Apache Spark'in temel bir yapisidir.
* Spark, mevcut koleksiyonu SparkContext paralellestirme yontemiyle islemek icin bir RDD'ye paralellestirerek verileri yukler. Veriler bir RDD'ye yuklendikten sonra bellekteki RDD'ler uzerinde donusumler ve eylemler (transformations) gerceklestirilir. Action sinyali gelinceye kadar bellekteki bu islemler devam etmektedir.

<br>



<br>

* _Spark, sistem hafizasi bitmedikce veya kullanici verileri kaliciligi saglamak icin diske yazmaya karar vermedikce verileri bellekte depolar._
* RDD'deki her veri kumesi, kumenin farkli dugumlerinde hesaplanabilen mantiksal bolumlere bolunmustur.
* Kullanicilar iki tur RDD islemi gerceklestirebilir: _Transformations & Actions_
* Transformations, yeni bir RDD olusturmak icin uygulanan islemlerdir.
* Actions, Apache Spark'a hesaplama uygulamasi ve sonucu surucuye geri gonderme talimatini vermek icin kullanilir.



<br>



<br>

Bu asamada gercek hayatta karsilasilabilecek baslangic icin uygun bir proje gerceklestirecegiz. Proje kullanacagimiz veriseti _Churn_ verisetidir. Bu veriseti icerisinde musteri bilgileri ve musterinin bizi terk edip etmeyecegini belirten _churn_ degiskeni bulunmaktadir.

Oncelikle findspark ile Spark'in konumunu belirtmemiz gerekiyor.

```python
import findspark
findspark.init("/home/ea/spark/spark-3.1.2-bin-hadoop3.2/")
```

PySpark uzerinden Session baslatma islemlerini gerceklestiriyoruz.

```python
import pyspark
from pyspark.sql import SparkSession
from pyspark.conf import SparkConf
from pyspark import SparkContext

spark = SparkSession.builder \
        .master("local") \
        .appName("churn_app") \
        .config("spark.executer.memory","16gb") \
        .getOrCreate()

sc = spark.sparkContext
sc
```

Spark'i baslattiktan sonra verisetimizin okuma islemini gerceklestiriyoruz. Bu islemi gerceklestirirken _header = True_ ile verisetinin column isimlerini df icerisine aktariyoruz, _inferSchema = True_ ile verisetindeki gozlem degerleri direkt olarak String alinmaz, Spark veriseti uzerinde gezinerek semasini cikartir ve buna gore df olusturur. Daha sonra _cache_ fonksiyonu ile df'i memorye tasiyoruz.

```python
spark_df = spark.read.csv("csv/churn.csv",
                         header=True,
                         inferSchema=True,
                         sep=",")
spark_df.cache()
```

Goruldugu uzere _c0_ olarak gozuken degisken degeri String olarak degil integer olarak tanimlanmistir.

```python
spark_df.printSchema()
```

DataFrame'e goz attigimizda column isimlerinin belirli bir duzende bulunmadigini goruyoruz. Standard olmasi acisindan tum harfleri buyuk ya da kucuk olarak kullanmak daha verimli olacaktir.

```python
spark_df.show(5)
```

Column isimlerini stardardize etmek icin _toDF_ fonksiyonunu kullanarak yeni column isimlerini list comprehension ile duzeltiyoruz.

```python
spark_df = spark_df.toDF(*[c.lower() for c in spark_df.columns])
spark_df.show(5)
```

DataFrame'e gozattigimizda c0 column degerinin indexleri ifade ettigini gormekteyiz. Bu sebeple yeniden isimlendirmek verisetinin kullanimini ve anlasilirligini kolaylastiracaktir.

```python
spark_df = spark_df.withColumnRenamed("_c0","index")
spark_df.show(1)
```

Veriseti hakkindaki temel birkac bilgi edinmek adina veriseti uzerinde basit fonksiyonlar calistirip fikir sahibi olmaya calsiiyoruz.

900 adet gozlem degerinden olusuyor.

```python
spark_df.count()
```

8 adet degisken degeri barindiriyor.

```python
print(len(spark_df.columns))
spark_df.columns
```

Her bir gozlem degeri birbiriyle cakismiyor, her biri birbirininden farkli durumdadir.

```python
spark_df.distinct().count()
```

Cakisma durumunu musteri isimlerine indirgedigimizde 2 adet gozlemin cakistigini gormekteyiz. 

```python
spark_df.select("names").distinct().count()
```

Bu cakismaya neden olan gozlemlere gozatarak duplicated bir veri olup olmadigini soyleyebiliriz.

Burada goruldugu uzere _Jennifer Wood_ isimli gozlemden iki adet bulundugu gorulmektedir.

```python
spark_df.groupby("names").count().sort("count",ascending=False).show(3)
```

Jennifer Wood names degerine sahip olan degiskenleri gozlemlememiz gerekiyor.

Goruldugu uzere iki gozlem de birbirinden farkli degerler icermektedir, bu sebeple hataya neden olacak bir cakisma soz konusu degildir.

```python
spark_df.filter(spark_df.names == "Jennifer Wood").show()
```

index degerini referans alarak bir gozleme ulasmak istedigimizde ise _where_ fonksiyonunu filterdaki benzer mantikla kullanabiliriz.

```python
spark_df.where(spark_df.index == 439).show()
```

Eger yukaridaki gibi elde ettigimiz bir gozlem degerinin _names_ gibi spesifik bir degiskenini _islenebilir olarak_ elde etmek istiyorsak bu islemi de **collect** fonksiyonu yardimiyla yapiyoruz.

```python
spark_df.where(spark_df.index == 439).collect()[0]
```

```python
spark_df.where(spark_df.index == 439).collect()[0]["names"]
```
